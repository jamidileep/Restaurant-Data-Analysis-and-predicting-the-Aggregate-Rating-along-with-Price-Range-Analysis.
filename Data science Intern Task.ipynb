# Predictive Modeling

```python
y=new_df['Aggregate rating']
x=new_df.drop(columns=['Aggregate rating'])
```

```python
y
```

```python
x
```

Split the dataset into training and testing sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
modelD = DecisionTreeRegressor(random_state = 0)
modelD.fit(X_train,y_train)
```

```python
y_predict=modelD.predict(X_test)
```

```python
y_predict
```

```python
y_predict_array = np.array(y_predict)
y_test_array = np.array(y_test)
y_predict_reshaped = y_predict_array.reshape(len(y_predict_array), 1)
y_test_reshaped = y_test_array.reshape(len(y_test_array), 1)
concatenated_arrays = np.concatenate((y_predict_reshaped, y_test_reshaped), axis=1)
concatenated_arrays
```

```python
print(concatenated_arrays)
```

Evaluate The Model Performance

```python
mse = mean_squared_error(y_test, y_predict)
mse
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_predict)
r2
```

```python
from sklearn.ensemble import RandomForestRegressor
modelR=RandomForestRegressor(n_estimators=100, random_state=0)
modelR.fit(X_train,y_train)
```

```python
y_predictR=modelR.predict(X_test)
y_predictRT=modelR.predict(X_train)
```

```python
mse = mean_squared_error(y_test, y_predictR)
mseT = mean_squared_error(y_train, y_predictRT)
```

```python
mse
```

```python
mseT
```

```python
r2 = r2_score(y_test, y_predictR)
r2T=r2_score(y_train, y_predictRT)
r2
```
